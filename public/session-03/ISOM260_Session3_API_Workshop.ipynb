## 📐 Step 3: Apply the CRAFT Framework

Now let's create a reusable function that applies the CRAFT framework:
- **C**ontext
- **R**ole
- **A**ction
- **F**ormat
- **T**one

```python
def craft_prompt(context: str, role: str, action: str, format: str, tone: str) -> str:
    """
    Build a well-structured prompt using the CRAFT framework.

    Args:
        context: Background information and situation
        role: Who the AI should act as
        action: What the AI should do
        format: How the output should be structured
        tone: The voice and style to use

    Returns:
        A formatted prompt string
    """
    prompt = f"""[CONTEXT]
{context}

[ROLE]
{role}

[ACTION]
{action}

[FORMAT]
{format}

[TONE]
{tone}"""

    return prompt

print("✅ CRAFT function created!")
```

```python
# Example: Using CRAFT for a business email
email_prompt = craft_prompt(
    context="I'm a junior project manager who needs to ask my team lead for a deadline extension on a major deliverable. The original deadline is Friday, and I need until next Wednesday.",
    role="You are a professional communication coach who specializes in workplace emails.",
    action="Write an email requesting the deadline extension. Be honest about the challenges while demonstrating accountability.",
    format="Subject line + email body. Keep it under 150 words.",
    tone="Professional but warm. Confident, not apologetic."
)

print("📧 Generated CRAFT Prompt:")
print("-" * 50)
print(email_prompt)
```

```python
# Send it to Claude!
response = client.messages.create(
    model="claude-sonnet-4-5-20250514",
    max_tokens=1024,
    messages=[{"role": "user", "content": email_prompt}]
)

print("📨 Claude's Email:")
print("=" * 50)
print(response.content[0].text)
```

### 📧 Option A: Email Tone Analyzer

```python
def analyze_email_tone(email_text: str) -> str:
    """
    Analyze the tone of an email and provide suggestions for improvement.
    """
    prompt = craft_prompt(
        context=f"I received or wrote this email and want to understand how it might be perceived:\n\n{email_text}",
        role="You are an expert communication analyst who specializes in workplace dynamics and emotional intelligence.",
        action="Analyze this email's tone. Identify: 1) The primary emotional tone, 2) How the recipient likely feels reading it, 3) Any phrases that could be misinterpreted, 4) Specific suggestions to make it more effective.",
        format="Use clear headers for each section. Be specific with examples from the email.",
        tone="Constructive and helpful. Like a trusted colleague giving feedback."
    )

    response = client.messages.create(
        model="claude-sonnet-4-5-20250514",
        max_tokens=1024,
        messages=[{"role": "user", "content": prompt}]
    )

    return response.content[0].text

# Test it!
sample_email = """
Subject: Project Update

Hi team,

I noticed that several of you haven't submitted your status reports yet.
This is the third time I've had to send a reminder.
The deadline was yesterday.

Please submit by EOD or let me know what the issue is.

Thanks,
Alex
"""

print("📊 Email Tone Analysis:")
print("=" * 50)
print(analyze_email_tone(sample_email))
```

### 📝 Option B: Meeting Notes Summarizer

```python
def summarize_meeting(raw_notes: str, attendees: str = "team") -> str:
    """
    Transform messy meeting notes into a structured summary with action items.
    """
    prompt = craft_prompt(
        context=f"These are raw notes from a meeting with {attendees}:\n\n{raw_notes}",
        role="You are an executive assistant known for creating crystal-clear meeting summaries that drive action.",
        action="Create a structured meeting summary. Extract: 1) Key decisions made, 2) Action items with owners and deadlines, 3) Open questions or parking lot items, 4) A one-paragraph executive summary.",
        format="Use markdown headers. Action items should be checkboxes with [Owner - Deadline] format.",
        tone="Professional and concise. Every word should earn its place."
    )

    response = client.messages.create(
        model="claude-sonnet-4-5-20250514",
        max_tokens=1500,
        messages=[{"role": "user", "content": prompt}]
    )

    return response.content[0].text

# Test it!
sample_notes = """
Q1 Planning Meeting - Jan 15

talked about the new product launch. sarah thinks we should push to march.
mike disagrees, wants to keep february date.
budget discussion - we're over by 15% on marketing spend
need to hire 2 more engineers - james will post job descriptions by friday
competitor analysis shows we're losing ground in enterprise segment
decided to focus on SMB for now, revisit enterprise in Q2
action: lisa to prepare customer research report
action: whole team to review OKRs before next meeting
next meeting: jan 22
someone mentioned security audit - not sure if urgent
sarah OOO next week
"""

print("📋 Meeting Summary:")
print("=" * 50)
print(summarize_meeting(sample_notes, "Product Team (Sarah, Mike, James, Lisa)"))
```

### 🛍️ Option C: Product Description Generator

```python
def generate_product_description(product_name: str, features: list, target_audience: str, price_point: str = "mid-range") -> str:
    """
    Generate a compelling product description from features.
    """
    features_text = "\n".join([f"- {feature}" for feature in features])

    prompt = craft_prompt(
        context=f"Product: {product_name}\nTarget Audience: {target_audience}\nPrice Point: {price_point}\n\nFeatures:\n{features_text}",
        role="You are a conversion copywriter who has written product descriptions for Apple, Nike, and leading D2C brands. You understand how to turn features into benefits.",
        action="Write a product description that: 1) Opens with an attention-grabbing headline, 2) Focuses on benefits over features, 3) Creates emotional connection with the target audience, 4) Includes a compelling call-to-action.",
        format="Headline (under 10 words) + 2-3 paragraphs + CTA. Total under 200 words.",
        tone="Confident and aspirational. Make the reader feel like this product will improve their life."
    )

    response = client.messages.create(
        model="claude-sonnet-4-5-20250514",
        max_tokens=1024,
        messages=[{"role": "user", "content": prompt}]
    )

    return response.content[0].text

# Test it!
description = generate_product_description(
    product_name="FocusPod Pro",
    features=[
        "Noise-canceling technology blocks 95% of ambient sound",
        "8-hour battery life",
        "Comfortable memory foam ear cushions",
        "Built-in microphone for calls",
        "Folds flat for travel",
        "Works with all Bluetooth devices"
    ],
    target_audience="Remote workers and students who need to focus in noisy environments",
    price_point="premium ($199)"
)

print("🛍️ Product Description:")
print("=" * 50)
print(description)
```

## 🔥 Step 5: Advanced Techniques

### Using System Prompts

System prompts set persistent behavior for your AI. They're like giving Claude a job description before the conversation starts.

```python
def chat_with_persona(user_message: str, system_prompt: str) -> str:
    """
    Chat with Claude using a custom system prompt (persona).
    """
    response = client.messages.create(
        model="claude-sonnet-4-5-20250514",
        max_tokens=1024,
        system=system_prompt,  # This is where the system prompt goes!
        messages=[{"role": "user", "content": user_message}]
    )

    return response.content[0].text

# Example: A financial advisor persona
financial_advisor_prompt = """
You are a certified financial planner with 20 years of experience.
You specialize in helping young professionals build wealth.
You always:
- Explain concepts in simple terms without jargon
- Provide balanced perspectives on risk
- Remind users you can't give specific investment advice
- Encourage long-term thinking over get-rich-quick schemes
"""

question = "I just got my first job and have $500/month to save. What should I do with it?"

print("💰 Financial Advisor Response:")
print("=" * 50)
print(chat_with_persona(question, financial_advisor_prompt))
```

### Temperature: Controlling Creativity

Temperature controls how "creative" or "deterministic" the AI is:
- `temperature=0`: Very consistent, factual (good for data analysis)
- `temperature=1`: More creative, varied (good for brainstorming)

```python
def generate_with_temperature(prompt: str, temperature: float) -> str:
    """Generate a response with specified temperature."""
    response = client.messages.create(
        model="claude-sonnet-4-5-20250514",
        max_tokens=200,
        temperature=temperature,
        messages=[{"role": "user", "content": prompt}]
    )
    return response.content[0].text

creative_prompt = "Give me a creative name for a coffee shop that's also a bookstore."

print("🧊 Temperature 0 (Consistent):")
print(generate_with_temperature(creative_prompt, 0))
print("\n" + "="*50 + "\n")
print("🔥 Temperature 1 (Creative):")
print(generate_with_temperature(creative_prompt, 1))
```

### Multi-Turn Conversations

Real applications often need back-and-forth conversations. Here's how to maintain context:

```python
class ConversationBot:
    """
    A simple chatbot that maintains conversation history.
    """
    def __init__(self, system_prompt: str = ""):
        self.system_prompt = system_prompt
        self.messages = []

    def chat(self, user_message: str) -> str:
        # Add user message to history
        self.messages.append({"role": "user", "content": user_message})

        # Make API call with full history
        response = client.messages.create(
            model="claude-sonnet-4-5-20250514",
            max_tokens=1024,
            system=self.system_prompt,
            messages=self.messages
        )

        assistant_message = response.content[0].text

        # Add assistant response to history
        self.messages.append({"role": "assistant", "content": assistant_message})

        return assistant_message

    def clear_history(self):
        self.messages = []

# Create a helpful coding tutor
tutor = ConversationBot(
    system_prompt="You are a patient Python tutor. Explain concepts simply and give small examples. Always encourage the student."
)

print("🎓 Multi-turn Conversation Demo:")
print("=" * 50)
print("\nUser: What is a list in Python?")
print(f"\nTutor: {tutor.chat('What is a list in Python?')}")

print("\n" + "-"*50)
print("\nUser: Can you show me how to add something to it?")
print(f"\nTutor: {tutor.chat('Can you show me how to add something to it?')}")
```

## 📊 Step 6: Cost Awareness

API calls cost money! Let's understand how to estimate and track costs.

```python
def estimate_cost(input_tokens: int, output_tokens: int, model: str = "claude-sonnet-4-5") -> dict:
    """
    Estimate the cost of an API call.

    Prices as of early 2025 (check anthropic.com for current pricing):
    - Claude Sonnet 4.5: $3/M input, $15/M output
    - Claude Haiku 4.5: $0.80/M input, $4/M output
    """
    pricing = {
        "claude-sonnet-4-5": {"input": 3.00, "output": 15.00},
        "claude-haiku-4-5": {"input": 0.80, "output": 4.00},
    }

    prices = pricing.get(model, pricing["claude-sonnet-4-5"])

    input_cost = (input_tokens / 1_000_000) * prices["input"]
    output_cost = (output_tokens / 1_000_000) * prices["output"]
    total_cost = input_cost + output_cost

    return {
        "input_tokens": input_tokens,
        "output_tokens": output_tokens,
        "input_cost": f"${input_cost:.6f}",
        "output_cost": f"${output_cost:.6f}",
        "total_cost": f"${total_cost:.6f}"
    }

# Calculate cost of our previous API call
cost = estimate_cost(
    input_tokens=message.usage.input_tokens,
    output_tokens=message.usage.output_tokens
)

print("💰 Cost Breakdown:")
for key, value in cost.items():
    print(f"   {key}: {value}")

print("\n💡 Tip: For development and testing, use Claude Haiku - it's much cheaper!")
```
